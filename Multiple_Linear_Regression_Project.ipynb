## Import The Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score,root_mean_squared_error
```

```python
dataset=pd.read_csv("AmesHousing.csv")
```

## Handling Missing Values

```python
# Categorical Features
dataset["Pool QC"] = dataset["Pool QC"].fillna("No Pool")
dataset["Misc Feature"] = dataset["Misc Feature"].fillna("None")
dataset["Alley"] = dataset["Alley"].fillna("No Alley")
dataset["Fence"] = dataset["Fence"].fillna("No Fence")
dataset["Mas Vnr Type"] = dataset["Mas Vnr Type"].fillna("None")
dataset["Fireplace Qu"] = dataset["Fireplace Qu"].fillna("No Fireplace")
dataset["Garage Qual"] = dataset["Garage Qual"].fillna("No Garage")
dataset["Garage Cond"] = dataset["Garage Cond"].fillna("No Garage")
dataset["Garage Finish"] = dataset["Garage Finish"].fillna("No Garage")
dataset["Garage Type"] = dataset["Garage Type"].fillna("No Garage")
dataset["Bsmt Exposure"] = dataset["Bsmt Exposure"].fillna("No Basement")
dataset["BsmtFin Type 1"] = dataset["BsmtFin Type 1"].fillna("No Basement")
dataset["BsmtFin Type 2"] = dataset["BsmtFin Type 2"].fillna("No Basement")
dataset["Bsmt Qual"] = dataset["Bsmt Qual"].fillna("No Basement")
dataset["Bsmt Cond"] = dataset["Bsmt Cond"].fillna("No Basement")
# Numerical Features
zero_fill_columns = [
    "Mas Vnr Area",
    "Bsmt Full Bath",
    "Bsmt Half Bath",
    "Total Bsmt SF",
    "BsmtFin SF 1",
    "BsmtFin SF 2",
    "Bsmt Unf SF",
    "Garage Area",
    "Garage Cars"
]
dataset[zero_fill_columns] = dataset[zero_fill_columns].fillna(0)
dataset["Electrical"] = dataset["Electrical"].fillna(dataset["Electrical"].mode()[0])


dataset.loc[dataset["Garage Type"] == "No Garage", "Garage Yr Blt"] = 0


dataset["Garage Yr Blt"] = dataset["Garage Yr Blt"].fillna(dataset["Year Built"])


dataset["Lot Frontage"] = dataset["Lot Frontage"].fillna(dataset["Lot Frontage"].median())

```

```python
dataset.isnull().sum().sum() 
# Now We Did handle the all missing values
```

### Feature Selection

```python
X = dataset.drop("SalePrice", axis=1)
y = dataset["SalePrice"]
```

## Encoding 

```python
categorical_columns = X.select_dtypes(include=["object", "string"]).columns
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(sparse_output=False),categorical_columns)],remainder='passthrough')
X = ct.fit_transform(X)
X = np.array(X)
print(X)
```

## Spliting The Teain set and Test set

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

## Training the Model 

```python
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

## Predicting the Test set Result

```python
y_pred=regressor.predict(X_test)
print(y_pred)
```
